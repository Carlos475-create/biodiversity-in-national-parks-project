### Import first dataset

```python
species = pd.read_csv('./species_info.csv')
species.head()
```

```python
species.info()
```

# Null values

```python
species.isnull().sum()/len(species)
```

```python
species['conservation_status'].unique()
```

After sampling a few records classified as `NaN`, it can be concluded that these missing values represent species with a **Least Concern** conservation status. Therefore, `NaN` values in this column will be replaced with `"Least Concern"` using `fillna()`.

```python
species['conservation_status'] = species['conservation_status'].fillna('Least Concern')
```

```python
species.isna().sum()
```
